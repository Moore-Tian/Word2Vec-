```python
from gensim.models import Word2Vec

# 载入降采样模型
model = Word2Vec.load("model/downsampling.model")
```

```python
model.wv.key_to_index # 获得所有词汇组
```

```python
model.wv.vectors.shape # 查看词向量大小
```

```python
model.wv['的'] # 得到词的词向量
```

```python
model.wv.similarity('中国', '美国') # 计算两个词的余弦相似度
```

```python
model.wv.most_similar('中国') # 取出与“中国”最相似的10个词
```

```python
model.wv.most_similar(positive=["中国", "纽约"], negative=["上海"]) # 获得 上海-中国+纽约 的词
```

```python
model.wv.closer_than("上海", "中国") # 比 中国 更接近 上海 的词汇
```
